```
#create Dataframe from Dictionary
import pandas as pd
data = {
    'Topic': ['Python', 'SQL', 'Pandas'],
    'Progress': [90, 85, 10],
    'Priority': ['High', 'High', 'Very High']
}
df = pd.DataFrame(data)
print(df)
```

```
# Create a simple DataFrame
sydney_flight = {
    'Day': [1, 2, 3, 4, 5],
    'Study_Hours': [4, 6, 5, 8, 7],
    'Mood': ['Happy', 'Focused', 'Tired', 'Excited', 'Strong']
}

df_trip = pd.DataFrame(sydney_flight)

# Challenge: Find data of study_hours more than 6
productive_days = df_trip[df_trip['Study_Hours'] > 6]

print("My Most Productive Days:")
print(productive_days)
```
