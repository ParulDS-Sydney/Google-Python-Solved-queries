```
#create Dataframe from Dictionary
import pandas as pd
data = {
    'Topic': ['Python', 'SQL', 'Pandas'],
    'Progress': [90, 85, 10],
    'Priority': ['High', 'High', 'Very High']
}
df = pd.DataFrame(data)
print(df)
```

```
# Create a simple DataFrame
sydney_flight = {
    'Day': [1, 2, 3, 4, 5],
    'Study_Hours': [4, 6, 5, 8, 7],
    'Mood': ['Happy', 'Focused', 'Tired', 'Excited', 'Strong']
}

df_trip = pd.DataFrame(sydney_flight)

# Challenge: Find data of study_hours more than 6
productive_days = df_trip[df_trip['Study_Hours'] > 6]

print("My Most Productive Days:")
print(productive_days)
```

```
# Data cleaning or Data wrangling

# Check for missing values
print(df.isnull()) # shows true/false in whole table

# Count missing values per column (Most Used!)
print(df.isnull().sum())
```

```
#how to handle missing data#

# Method 1:delete rows having missing values
df_cleaned = df.dropna()

# Method 2:any no. or anything can be replaced by missing values (e.g., 0)
df_filled = df.fillna(0)

# Method 3: Specific column having mean (average)
if 'Study_Hours' in df.columns:
    df['Study_Hours'] = df['Study_Hours'].fillna(df['Study_Hours'].mean())
    print("Missing values filled successfully!")
else:
    print("Error: Column name match nahi ho raha. List of columns:", df.columns)
```

```
# Dirty Data
import numpy as np
data = {
    'Name': ['Parul', 'Rohan', 'Parul', 'Amit'],
    'Score': [95, np.nan, 95, 80],
    'City': ['Sydney', 'Sydney', 'Sydney', 'Melbourne']
}

df_dirty = pd.DataFrame(data)

print("--- Dirty Data ---")
print(df_dirty)

# 1. Remove Duplicate
df_clean = df_dirty.drop_duplicates().copy()

# 2. Missing score is 0
df_clean['Score'] = df_clean['Score'].fillna(0)

print("\n--- Cleaned Data ---")
print(df_clean)
```

```
#sorted values

flights = {
    'Airline': ['Qantas', 'Emirates', 'Air India', 'Singapore'],
    'Price': [1200, 1500, 900, 1100],
    'Duration': [15, 20, 18, 16]
}

df_flights = pd.DataFrame(flights)

# Task: show cheapest flight
cheapest_flights = df_flights.sort_values(by='Price')

print("Flights sorted by Price:")
print(cheapest_flights)
```

```
#Data Filtering#

import pandas as pd

# Creating your prep data
data = {
    'Task': ['SQL Joins', 'Python Loops', 'Pandas Cleaning', 'Statistics', 'EDA'],
    'Days_Left': [2, 5, 4, 10, 8],
    'Status': ['Completed', 'Completed', 'In Progress', 'Pending', 'Pending']
}

df_prep = pd.DataFrame(data)

# 1. show only pending task
pending_tasks = df_prep[df_prep['Status'] == 'Pending']

# 2. show Pending task and more than 8 days are left
urgent_pending = df_prep[(df_prep['Status'] == 'Pending') & (df_prep['Days_Left'] > 8)]

print("Urgent Pending Tasks:")
print(urgent_pending)
```

```
#Data Grouping#

import pandas as pd

study_data = {
    'Subject': ['Python', 'Python', 'SQL', 'SQL', 'Pandas', 'Pandas'],
    'Source': ['YouTube', 'Course', 'YouTube', 'Practice', 'YouTube', 'Practice'],
    'Hours': [2, 4, 3, 5, 2, 6]
}

df_study = pd.DataFrame(study_data)

# Sawaal: Har Subject par total kitne ghante lagaye?
total_hours = df_study.groupby('Subject')['Hours'].sum()

# Sawaal: Har Source (YouTube vs Practice) se average kitni padhai hui?
source_efficiency = df_study.groupby('Source')['Hours'].mean()

print("Total Hours per Subject:")
print(total_hours)
print("\nAverage Hours per Source:")
print(source_efficiency)
```
