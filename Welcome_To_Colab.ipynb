```
#loop control continue#
flights = ["Confirmed", "Cancelled", "Confirmed", "Delayed", "Confirmed"]
confirmed_count = 0

for status in flights:
    if status == "Cancelled" or status == "Delayed":
        continue # Inhe skip karo
    confirmed_count += 1
    print(f"Processing flight {confirmed_count}: Ready to fly!")

print(f"\nTotal confirmed flights for Sydney: {confirmed_count}")
```

```
#Classes creating objects#

class Applicant:
    def __init__(self, name, role):
        self.name = name
        self.role = role
        self.status = "Pending"

    def update_status(self, new_status):
        self.status = new_status
        print(f"Status for {self.name} updated to: {self.status}")

# Creating "Objects" (Real-world examples of the class)
candidate1 = Applicant("Parul", "Data Scientist")
candidate2 = Applicant("Rohan", "Analyst")

candidate1.update_status("Shortlisted")

```

```
#Constructor: __init__ (The Entry Point)

class Student:
    # Constructor: It sets the initial data
    def __init__(self, name, target):
        self.name = name          # self matlab "is object ka"
        self.target = target

# Creating an Object
s1 = Student("Parul", "Google Sydney")

print(s1.name)   # Output: Parul
print(s1.target) # Output: Google Sydney
```

```
#Class Methods
class Student:
    def __init__(self, name, target):
        self.name = name
        self.target = target
        self.skills = []

    # Method to add skills
    def add_skill(self, new_skill):
        self.skills.append(new_skill)
        print(f"{new_skill} added to {self.name}'s profile!")

# Practical Use
p = Student("Parul", "Sydney")
p.add_skill("Python")
p.add_skill("SQL")

print(f"Current Skills: {p.skills}")
```

```
#Data Filtering#

import pandas as pd

# Creating your prep data
data = {
    'Task': ['SQL Joins', 'Python Loops', 'Pandas Cleaning', 'Statistics', 'EDA'],
    'Days_Left': [2, 5, 4, 10, 8],
    'Status': ['Completed', 'Completed', 'In Progress', 'Pending', 'Pending']
}

df_prep = pd.DataFrame(data)

# 1. show only pending task
pending_tasks = df_prep[df_prep['Status'] == 'Pending']

# 2. show Pending task and more than 8 days are left
urgent_pending = df_prep[(df_prep['Status'] == 'Pending') & (df_prep['Days_Left'] > 8)]

print("Urgent Pending Tasks:")
print(urgent_pending)
```

```
#Data Grouping#

import pandas as pd

study_data = {
    'Subject': ['Python', 'Python', 'SQL', 'SQL', 'Pandas', 'Pandas'],
    'Source': ['YouTube', 'Course', 'YouTube', 'Practice', 'YouTube', 'Practice'],
    'Hours': [2, 4, 3, 5, 2, 6]
}

df_study = pd.DataFrame(study_data)

# Sawaal: Har Subject par total kitne ghante lagaye?
total_hours = df_study.groupby('Subject')['Hours'].sum()

# Sawaal: Har Source (YouTube vs Practice) se average kitni padhai hui?
source_efficiency = df_study.groupby('Source')['Hours'].mean()

print("Total Hours per Subject:")
print(total_hours)
print("\nAverage Hours per Source:")
print(source_efficiency)
```
