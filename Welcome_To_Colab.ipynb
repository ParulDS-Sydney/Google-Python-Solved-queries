```
# Dictionary of our targets
sydney_roadmap = {
    "May": "Python Basics",
    "June": "SQL Advanced",
    "July": "Machine Learning"
}

# Goal: Print each month's focus
print("--- MY GOOGLE SYDNEY TIMELINE ---")
for month, task in sydney_roadmap.items():
    print(f"In {month}, my priority is: {task}")

# Bonus: List comprehension for success vibes
success_status = [f"{task} DONE!" for task in sydney_roadmap.values()]
print("\nFuture Status:", success_status)
```

```
#While loop#
savings = 0
goal = 500
days = 0

while savings < goal:
    savings += 100
    days += 1
    print(f"Day {days}: Saved ${savings}")

print("Goal Reached! Packing starts now.")
```

```
#loop control(break)#

data_logs = ["Success", "Success", "Error", "Success"]

for log in data_logs:
    if log == "Error":
        print("System Alert: Error found! Stopping check.")
        break
    print(f"Processing: {log}")
```

```
#Nested Loops (Loop inside a Loop)#
days = ["Day 1", "Day 2", "Day 3"]
spots = ["Opera House", "Bondi Beach"]

for d in days:
    print(f"--- Plan for {d} ---")
    for s in spots:
        print(f"Visiting: {s}")
```

```
#loop control continue#
flights = ["Confirmed", "Cancelled", "Confirmed", "Delayed", "Confirmed"]
confirmed_count = 0

for status in flights:
    if status == "Cancelled" or status == "Delayed":
        continue # Inhe skip karo
    confirmed_count += 1
    print(f"Processing flight {confirmed_count}: Ready to fly!")

print(f"\nTotal confirmed flights for Sydney: {confirmed_count}")
```

```
#Looping with Index (enumerate)

candidates = ["Riya", "Parul", "Amit", "Sana", "Deepak"]

for index, name in enumerate(candidates):
    if index < 3:
        print(f"Rank {index + 1}: {name} is shortlisted!")
```
